```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/kaggle/input/playground-series-s3e17/train.csv',index_col=0)
df
```

```python
Type = pd.get_dummies(df["Type"],drop_first=True)
print(Type)
```

```python
df1 = df.drop('Type',axis=1)
df1.head()
```

```python
df1 = pd.concat([df1,Type],axis=1)
df1
```

```python
df2 = df1.copy()
df2.head()
```

```python
df2['Product ID']=df2['Product ID'].str[1:]
```

```python
df2
```

```python
df_test = pd.read_csv('/kaggle/input/playground-series-s3e17/test.csv',index_col=0)
df_test.head()
```

```python
Type_test = pd.get_dummies(df_test["Type"],drop_first=True)
```

```python
df_test = df_test.drop('Type',axis=1)
df_test.head()
```

```python
df_test = pd.concat([df_test,Type_test],axis=1)
df_test
```

```python
df_test['Product ID']=df_test['Product ID'].str[1:]
```

```python
df_test
```

```python
df2['Machine failure'].value_counts()
```

```python
X = df2.drop("Machine failure", axis="columns")
y = df2["Machine failure"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=10000, random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold

lr = LogisticRegression(solver="liblinear")
params = {
    "C":[0.1, 1.0, 10.0],
    "penalty":["l1", "l2"]
}
cv = KFold(n_splits=5)
scoring = "roc_auc"
gs = GridSearchCV(lr, param_grid=params, cv=cv, scoring=scoring)
```

```python
gs.fit(X_train,y_train)
```

```python
gs.best_params_
```

```python
gs.best_score_
```

```python
y_test_pred = gs.predict(df_test)
y_test_proba = gs.predict_proba(df_test)
```

```python
df_test_id = df_test.index
X_test_ = pd.Series(df_test_id)
X_test_
```

```python
y_test_ = pd.Series(y_test_proba[:,1],name='Machine failure')
y_test_
```

```python
submission = pd.concat([X_test_,y_test_],axis=1)
```
